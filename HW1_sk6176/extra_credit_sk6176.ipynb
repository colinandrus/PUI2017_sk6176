```python
import os
import sys

# you can use the numpy library for random number generation: 
# http://docs.scipy.org/doc/numpy/reference/routines.random.html

import numpy as np
```

```python
np.random.seed(10)
```

```python
ReprRand=np.random.randn(2, 100)
```

```python
ReprRand.shape
```

```python
import pylab as pl
%pylab inline
```

## Fig.1 plot of ReprRand

```python
for i in range(100):
    pl.plot(ReprRand[0][i],ReprRand[1][i], 'o')
    plt.xlabel('first row of random number generator')
    plt.ylabel('second row of random number generator')
    plt.title('plot of ReprRand')
```

```python
ReprRandAll = np.zeros((50, 2, 100))
for i in range(50):
    new_mean = np.random.randn()
    sigma = np.random.randn()
    ReprRandAll[i] = sigma*ReprRand + new_mean
```

```python
ReprRandAll.shape
```

## Fig.2 plot of ReprRandAll

```python
for i in range(50):
    for j in range(100):
        pl.plot(ReprRandAll[i][0][j],ReprRandAll[i][1][j], 'x')
        plt.xlabel('first row of random number generator')
        plt.ylabel('second row of random number generator')
        plt.title('plot of ReprRandAll')
```
